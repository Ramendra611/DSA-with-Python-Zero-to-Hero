````python
## given a list of numbers, find the minimum of the numbers


nums = [4, 6, 8, 10]


# Method 1 : Using in built function
min_number = min(nums)
print(min_number)


# Method 2: Writing loop but still using in built function
min_number = nums[0]
for num in nums:
    min_number = min(min_number, num)
print(min_number)


# Method 3: Without using in-built function
min_number = nums[0]
for num in nums:
    if num < min_number:
        min_number = num
print(min_number)
````

````python
'''Given the list of marks in 3 subjects for students, create a mapping of student and total and 
average marks and the result. 
If the student fails in any subject, then overall he is fail. 


result_dict = {

    "Anubhav": {
                    marks:
                    total_marks : <>, 
                    average_marks : <>, 
                    maximum_marks : <>, 
                    minimum_marks : <>, 
                    result = <pass/fail>, 
                    fail_subjects: <>
                    }, 

    "Navneet":
}
'''

names_list = ['Anubhav', 'Navneet', 'Deviprasad']
marks_list = [[50, 60 , 90] , [30, 50, 80], [40, 100, 20]]
passing_marks = 40 
result_dict = {}
for ind, marks in enumerate(marks_list):
    # print(marks)
    total_marks = 0
    min_marks, max_marks = marks[0], marks[0]
    result = "pass" # flag
    for mark in marks:
        total_marks = total_marks + mark
        min_marks = min(min_marks, mark)
        max_marks = max(max_marks, mark)
        if mark < passing_marks:
            result = "fail"
        
    average_marks = total_marks / len(marks)

    # print(f"{total_marks= }, {min_marks = }, {max_marks = } {average_marks = }   ")

    result_dict[names_list[ind]] = {
        "marks": marks, 
        "total_marks": total_marks, 
        "average_marks" : average_marks, 
        "min_marks": min_marks, 
        "max_marks" : max_marks,  
        "result" : result
    }

    # print("--" *20)
print(result_dict)
````

## Dictionary comprehensions
    - it is alternate syntax to create key-value pairs

````python
numbers = [1, 2, 3, 4, 5]

result_dict = {
        1 : 1, 2:4, 3:9, 4:16, 5:25
    }



# traditional
result_dict = {}
for num in numbers:
    result_dict[num] = num ** 2 
print(result_dict)


## using dictionary comprehension
result_dict = { num : num ** 2  for  num in numbers }
print(result_dict)

````

````python
## Create a dictionary from the list of first name and last names 
first_names = [ "Debashish", "Siddhi", "Rajiv", "Anubhav"]
last_names = ["Choudhury", "Raj", "Kumar", "Pandey"]

names = {}
for first_name, last_name in zip(first_names, last_names):
    names[first_name] = last_name
print(names)

## dictionary comprehension

names = {f_name : l_name for (f_name, l_name) in zip(first_names, last_names) }
print(names)
````

````python
## Create a dictionary from the list of first name and last names where the key will be full name and value will be email id 
first_names = [ "Debashish", "Siddhi", "Rajiv", "Anubhav"]
last_names = ["Choudhury", "Raj", "Kumar", "Pandey"]

expected_result = {
    "Debashish Choudhury" : "debashish.choudhury@codeverra.com", 
    "Siddhi Raj" : "siddhi.raj@codeverra.com", 
    "Rajiv Kumar": "rajiv.kumar@codeverra.com"
}


## traditional approach
names_dict  = {}

for f_name, l_name in zip(first_names, last_names):
    # full_name = f_name + " "  + l_name # key
    full_name = f"{f_name} {l_name}"
    # email_id = f_name+"."+l_name+"@codeverra.com" # value
    email_id = f"{f_name}.{l_name}@codeverra.com ".lower()
    names_dict[full_name] = email_id

print(names_dict)

## dictionary comprehension

names_dict = { f"{f_name} {l_name}" : f"{f_name}.{l_name}@codeverra.com ".lower() for f_name, l_name  in zip(first_names, last_names)  }
print(names_dict)
````

````python
## Write a code to count the frequency of letters in a string

s = "hello"
# freq_count = { "h" : 1, "e": 1, "l": 2, "o" : 1 }


freq_count = {}
for char in s:
    ## get existing frequency
    if char in freq_count:
        existing_frequency  = freq_count[char]
    else:
        existing_frequency = 0
    # add one to existing frequency
    updated_frequency = existing_frequency + 1
    freq_count[char] = updated_frequency

print(freq_count)


freq_count = {}
for char in s:
    ## get existing frequency
    existing_frequency  = freq_count[char] if char in freq_count else 0
    # add one to existing frequency
    updated_frequency = existing_frequency + 1
    freq_count[char] = updated_frequency

print(freq_count)


freq_count = {}
for char in s: 
    freq_count[char] = freq_count.get(char, 0 ) + 1
print(freq_count)
````

## While Loops

    - for loop we know how many times we will iterate
    - while loop, we can iterate until a condition is met

````python
'''
Design a system where a user can enter the password and if the password 
is correct then display "logged in" otherwise ask the user to enter the password again
'''

correct_password = "5789"

entered_password = input("Please enter the password")

for i in range(3):
    if entered_password == correct_password:
        print("logged in")
    else:
        print("incorrect password. Please try again!!")
        entered_password = input("Please enter the password")

````

````python
## while loop

## Given a list of numbers, calculate the square of the numbers
numbers = [1, 2, 3, 4, 5, 6]

ind = 0
squared_nums = []
while ind < len(numbers): # loop will run untill the ind reaches the value 6
    print("the value of ind is: ", ind)
    num = numbers[ind]
    squared_nums.append(num ** 2)
    ind = ind + 1
    
print(squared_nums)
    
````

````python
'''
Design a system where a user can enter the password and if the password 
is correct then display "logged in" otherwise ask the user to enter the password again
'''

correct_password = "4789"
entered_password = input("Please enter the password")

while entered_password != correct_password: 
    # stop the while loop when entered_password becomes equal to correct_password
    if entered_password == correct_password:
        print("logged in")
    else:
        print("incorrect password. Please try again!!")
        entered_password = input("Please enter the password")
````
