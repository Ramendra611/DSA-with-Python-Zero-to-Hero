```python
# Question: Given a list, find the square of every number in the list

nums = [2, 3, 5, 9, 10]

square_list = []
for num in nums:
    # calculate the square of num
    square_of_num = num ** 2

    # put the square in the new list
    square_list.append(square_of_num)

print(nums)
print(square_list)
```

```python
# Question : Given a list of nums, separate out even and odd numbers and store it in a dictionary
'''
nums = [1, 2, 3, 4, 5, 6, 7, 8]
result = {
        "nums": [1, 2, 3, 4, 5, 6, 7, 8]
        "even": [2, 4, 5, 6, 8], 
        "odd": [1, 3, 5, 7]
}
'''
# nums = [20, 35, 56, 78, 101, 12]
nums= [1, 2, 3, 4, 5, 6, 7, 8]
# even_nums = []
# odd_nums = []


# initialise a dictionary with keys "even" and "odd"
result = {}  # dont initialise a dictionary with the keys. Keep the dictionary empty

for num in nums:

    if "nums" not in result:
        result["nums"] = []
    # store the number in the dictionary
    result["nums"].append(num)
    
    # check if num is even
    if num % 2 == 0 : # divide by 2 and remainder is 0
        
        if "even" not in result:
            result["even"] = []
        result["even"].append(num)
    else: # if the number is odd
        
        if "odd" not in result:
            result["odd"] = []
        result["odd"].append(num)

print(result)
```
