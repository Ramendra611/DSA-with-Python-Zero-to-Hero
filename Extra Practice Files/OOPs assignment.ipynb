## OOPs assignment 3 - Shape class

#### Create a Base Class
Define a Shape class with attributes name and methods:
        area() – returns 0 by default
        perimeter() – returns 0 by default

#### Write a method to display the shape’s name.

#### Square Class
Create a Square class that inherits from Shape.
Add an attribute side.
        Override the area() and perimeter() methods to calculate:
        
        area = side * side
        perimeter = 4 * side


#### Rectangle Class
Create a Rectangle class that inherits from Shape.
Add attributes length and width.
Override the area() and perimeter() methods.

#### Circle Class
Create a Circle class that inherits from Shape.
Add an attribute radius.
Override the area() and perimeter() methods using:

        area = π * radius^2
        perimeter = 2 * π * radius


#### Triangle Class
Create a Triangle class that inherits from Shape.
Add attributes side_a, side_b, side_c.

The perimeter() is the sum of the three sides.

Use Heron’s formula for area:

s = (side_a + side_b + side_c) / 2
area = √[s * (s - side_a) * (s - side_b) * (s - side_c)]

#### Polymorphism Demonstration
Create a list of different shape objects (Square, Rectangle, Circle, etc.).
Loop through the list and call area() and perimeter() on each shape to demonstrate polymorphism.

#### Input Validation
Add validation to ensure that shape dimensions (like side, length, radius) are positive numbers.
If not, display an error message.

#### Largest Area Finder
Write a function find_largest_area(shapes_list) that accepts a list of shape objects and returns the shape with the largest area.

#### Comparison Method
In the Shape class, add a method compare_area(other_shape) that compares the area of the current shape with another shape and prints which one has the larger area.

#### Advanced (Optional)
Add a ColoredShape subclass that inherits from any shape and includes an extra attribute color.
Override the display method to include both the shape’s details and its color.
