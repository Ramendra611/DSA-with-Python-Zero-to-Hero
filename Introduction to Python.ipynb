## Assignment Operatorsm

````python
x = 10
y = 20

print(x + y)   # Addition
print(x - y)   # Subtraction
print(x * y)   # Multiplication
print(x / y)   # Division
print(x // y)  # Floor Division --> Quotient
print(x % y)   # Modulo operator --> giving the remainder
print(2 ** 3)  # Exponential Operator


z = -x
print(z)  # negation
````

## Comparison Operators

````python
x = 10
y = 5
z = 10.0

# check if x and y are equal
print(x == y) ## == is for checking if two values are equal 
print(x == z)

print( x != y ) ## x is not equal to y


print("x is greater than y" , x > y)
print("x is smaller than y", x < y)

print(x >= y)
print(x <= y)


print(" chaining two comparisons ", 10 < 50 < 100)
print(" chaining two comparisons ", 100 < 50 < 200)
````

## Logical Operators

````python
x = 5
y = 2

print(x > 3 and y < 3) 
print(x > 3 and y < 1)

print(x > 3 or y < 1)

print(not x > 3)
````

## Membership Operator

using python```in``` and python```not in```

````python
name = "Rajiv"

print("j" in name) # in is a membership operator

print("p" in name)


print("ashish" in "Debashish")


list_of_students = [ "Rajiv", "Manya", "Siddhi" ]
name = "Siddhi"
name = "Ramesh"
print("Membership of ", name , name in list_of_students)
````

## Identity Operator
    - used to check if two items are the same object

````python
a = "Ramesh"
b = "Suresh"
c = a
d  = "Ramesh"

print(a == d) # equality
print(a is d) # identity

print(a is b)
print(c is a)
````

````python
id(a), id(d)
````
