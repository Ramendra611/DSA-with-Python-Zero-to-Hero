```python
# Conditional statements

## Take a decision based on a condition
shabana_marks = 30

passing_marks = 33

shabana_marks > passing_marks

```

```python
if (shabana_marks > passing_marks):
    print("Pass!!")
else:
    print("Fail")
```

## Ask the user to enter his age and based on the age tell what is to be done
    - age 0-17 ---> study
    - age 17- 40 ---> work
    - age 40-60 ---> relax
    - age > 60 ----> pray to god

```python
age = 70

if age < 17:
    print("Please study!")
else:
    if age < 40: 
        print("Please focus on work!!")
    else:
        if age < 60:
            print("You can relax now!!")

        else:
            print("You should pray to God!!!")
```

```python
## if elif else

age = 40

if age < 17:
    print("please study")
elif age < 40:  # else + if
    print("Please focus on work!!")
elif age < 60:
    print("You can relax now!!")
else:
    print("You should pray to God!!!")
```

```python
## we can use the input function to get input from user

age = input("Please enter your age ( in integer)")
# age = int(age)
age = float(age)

if age < 17:
    print("please study")
elif age < 40:  # else + if
    print("Please focus on work!!")
elif age < 60:
    print("You can relax now!!")
else:
    print("You should pray to God!!!")
    
```

```python
age = input("Please enter your age ( in integer)")
# age = int(age)
age = float(age)

if age < 0:
    print("Please enter a valid age !")

if age > 0 and age <= 17:
    print("please study")
elif age > 17 and age < 40:  # else + if
    print("Please focus on work!!")
elif age < 60:
    print("You can relax now!!")
else:
    print("You should pray to God!!!")
```
